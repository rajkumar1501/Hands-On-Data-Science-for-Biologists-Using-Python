## Importing the libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
mpl.rcParams['figure.dpi'] = 300
```

## Importing the dataset

```python
dataset = pd.read_csv('heart.csv')
dataset.head()
```

### Features:
<ol>
<li>age</li>
<li>sex</li>
<li>chest pain type (4 values)</li>
<li>resting blood pressure</li>
<li>serum cholestoral in mg/dl</li>
<li>fasting blood sugar &gt; 120 mg/dl</li>
<li>resting electrocardiographic results (values 0,1,2)</li>
<li>maximum heart rate achieved</li>
<li>exercise induced angina</li>
<li>oldpeak = ST depression induced by exercise relative to rest</li>
<li>the slope of the peak exercise ST segment</li>
<li>number of major vessels (0-3) colored by flourosopy</li>
<li>thal: 3 = normal; 6 = fixed defect; 7 = reversable defect</li>
</ol>

```python
dataset['target'].value_counts()
```

```python
sns.pairplot(dataset[['age','trestbps','chol','thalach','oldpeak','target']],hue='target')
```

## Checking if there is any null value

```python
dataset.isnull().sum()
```

```python

X = dataset.iloc[:,:-1]
print(X.info())
y = dataset.iloc[:,-1]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
#To retrieve the intercept:
print(classifier.intercept_)
#For retrieving the slope:
print(classifier.coef_)
```

```python
coeff_df = pd.DataFrame(classifier.coef_.T, X.columns, columns=['Coefficient'])  

coeff_df  =coeff_df.sort_values(by=['Coefficient'],ascending=False)
coeff_df
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
sns.heatmap(cm,annot=True,fmt='d',cmap="gray")
```

## Evaluation

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## Case study: Breast Cancer Wisconsin (Diagnostic) Data Set

```python
dataset = pd.read_csv('Breast_Cancer.csv')
dataset.head().T
```

```python
dataset.shape
```

```python
dataset['diagnosis'].value_counts()
```

```python
dataset = dataset.drop(['id','Unnamed: 32'],axis=1)
```

```python
X = dataset.iloc[:,1:]
y = dataset.iloc[:,0]
```

## Data Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_norm = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size = 0.2, random_state = 0)
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)

print(classification_report(y_test, y_pred))
```

```python
coeff_df = pd.DataFrame(classifier.coef_.T, X.columns, columns=['Coefficient'])  

coeff_df  =coeff_df.sort_values(by=['Coefficient'],ascending=False)
coeff_df
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10,4))
sns.boxplot(x="radius_se", y="diagnosis", data=dataset,ax=axes[0])
sns.boxplot(x='texture_worst',y='diagnosis',
            data=dataset,ax=axes[1])
fig.tight_layout()
```
